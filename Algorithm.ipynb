Depth First Search

```python
def DFS(graph, v, visited):
  visited[v] = True
  print(v)
  for node in graph[v]: # 이 부분이 stack에 쌓아서 방문 -> 재귀적으로 처리
    if not visited[node]:
      DFS(graph, node, visited)
graph = [[],
         [2, 3, 8],
         [1, 7],
         [1, 4, 5],
         [3, 5],
         [3, 4],
         [7],
         [2, 6, 8],
         [1, 7]
]         
visited = [False] * 9
DFS(graph, 1, visited)
```

Breadth First Search

```python
from collections import deque
def BFS(graph, v, visited):
  queue = deque([v])
  visited[v] = True
  while queue:
    value = queue.popleft()
    print(value)
    for i in graph[value]:
      if not visited[i]:
        queue.append(i)
        visited[i] = True

graph = [[],
         [2, 3, 8],
         [1, 7],
         [1, 4, 5],
         [3, 5],
         [3, 4],
         [7],
         [2, 6, 8],
         [1, 7]
]         
visited = [False] * 9
BFS(graph, 1, visited)
```

```python
# 덧칠하기 
from collections import deque
def solution(n, m, section):
    answer = 0
    section = deque(section)
    while section:
        start = section.popleft()
        while section:
            if section[0] >= start + m:
                break
            section.popleft()
        answer += 1
    return answer
```

#### 연속펄스 부분수열

```python
# 연속펄스 부분수열
# 펄스 수열이란 [1, -1, 1, -1 …] 또는 [-1, 1, -1, 1 …]
# sequence가 주어졌을 때 위 펄스 수열 두개 모두 sequence와 곱해서
# 펄스 전체 수열을 얻고 거기서 부분을 선택하여 최댓값을 찾는 문제
# 틀린 답 + 시간 초과

sequence = [2, 3, -6, 1, 3, -1, 2, 4]

def solution(sequence):
    table = [[0 for _ in range(len(sequence) + 1)] for _ in range(2)]
    weight = 1
    for i in range(len(sequence)):
        # sign을 판단하며 누적합을 구한다
        # i~j까지 최대값을 구하는 문제이므로
        # 부분합이 i일때 가장 크고, j일때 가장 작은 페어를 찾는다
        # i와 j사이의 부분합 차이를 구하면 최대값
        table[0][i + 1] = table[0][i] + sequence[i] * weight
        table[1][i + 1] = table[1][i] - sequence[i] * weight
        weight *= -1

    print(table)
    return max(max(table[0]) - min(table[0]), max(table[1]) - min(table[1]))

solution(sequence)
```
